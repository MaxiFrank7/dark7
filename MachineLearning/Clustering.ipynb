PROYECTO 3: Segmentaci√≥n de Clientes
Escenario: Marketing te da una base de datos con Edad y Puntaje_de_Gasto (1-100) de usuarios de un centro comercial. Quieren lanzar campa√±as publicitarias espec√≠ficas, pero no saben cu√°ntos tipos de clientes hay.

üõ†Ô∏è Paso 1: Generaci√≥n de Datos (Copia y Pega)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Generamos datos con 5 grupos naturales ocultos
from sklearn.datasets import make_blobs

# data[0] son las caracter√≠sticas, data[1] son las etiquetas (que ignoraremos)
datos, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

df_mall = pd.DataFrame(datos, columns=['Edad_Normalizada', 'Puntaje_Gasto'])

# Visualicemos los datos "crudos" (lo que ve el algoritmo)
plt.scatter(df_mall['Edad_Normalizada'], df_mall['Puntaje_Gasto'])
plt.title("Clientes sin Etiquetar")
plt.xlabel("Edad")
plt.ylabel("Gasto")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
#entrena el modelo, se pasa df entero
kmeans.fit(df_mall)
#predice
etiquetas=kmeans.predict(df_mall)
#visualizacion
plt.scatter(df_mall['Edad_Normalizada'], df_mall['Puntaje_Gasto'],c=etiquetas,cmap='rainbow')
plt.title("Segmentaci√≥n de Clientes Detectada por IA")
plt.xlabel("Edad (Normalizada)")
plt.ylabel("Puntaje de Gasto")
plt.show()
```
