```python
"""Importa LinearRegression de sklearn.linear_model.

1)Prepara los datos:

Variable X: Columna Años_Experiencia (¡Ojo con el doble corchete!).
Variable y: Columna Salario.

2) Crea el modelo y entrénalo (.fit(X, y)).

3)Predicción: ¿Cuánto debería cobrar alguien con 10 años de experiencia?

Nota: Para predecir, el dato nuevo también debe estar en formato 2D. Usa [[10]] dentro del predict."""

data = {
    'Nombre': ['Ana', 'Luis', 'Juan', 'Maria', 'Pedro', 'Laura'],
    'Departamento': ['Ventas', 'IT', 'Ventas', 'IT', 'Marketing', 'Ventas'],
    'Salario': [2500, 4500, 2700, 4800, 3200, 2600],
    'Años_Experiencia': [2, 8, 3, 9, 4, 2]
}
df_empresa = pd.DataFrame(data)

modelo=LinearRegression()

X=df_empresa[['Años_Experiencia']]
y=df_empresa[['Salario']]
modelo.fit(X, y)

prediccion=modelo.predict([[10]])

print(f"Salario estimado para 10 años: ${prediccion[0][0]:.2f}")
```
