```python
from xgboost import XGBRegressor
import pandas as pd
import numpy as np

# 1. Recuperamos la lógica del "Sandwich" (Dummies)
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
y_log = np.log1p(df_train['SalePrice'])

# Unir, Dummies y Separar (Como hicimos en Random Forest)
df_total = pd.concat([df_train.drop(['SalePrice', 'Id'], axis=1), df_test.drop(['Id'], axis=1)])
df_total_dummies = pd.get_dummies(df_total) # Convierte todo texto a numero
X_all = df_total_dummies.fillna(0)

X = X_all.iloc[:1460]
X_test_kaggle = X_all.iloc[1460:]

# 2. ENTRENAMIENTO CON XGBOOST
# n_estimators: Cuántos "alumnos" corrigiéndose hay en la fila
# learning_rate: Qué tanto caso le hago al alumno nuevo (0.05 es conservador y bueno)
modelo_xgb = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1)

print("Entrenando XGBoost...")
modelo_xgb.fit(X, y_log)

# 3. Predicción
pred_log_xgb = modelo_xgb.predict(X_test_kaggle)
pred_final_xgb = np.expm1(pred_log_xgb)

sub = pd.DataFrame({'Id': df_test['Id'], 'SalePrice': pred_final_xgb})
sub.to_csv('house_xgboost.csv', index=False)
print("¡Archivo 'house_xgboost.csv' listo!")
```
