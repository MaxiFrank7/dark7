```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

El Caso: "Inmobiliaria Inteligente" üè† Tienes datos de casas y quieres predecir su precio.

* 1) Genera un DataFrame con 100 casas:

Metros_Cuadrados: Aleatorio entre 50 y 200.

Habitaciones: Aleatorio entre 1 y 5.

Precio: Formula inventada: (Metros * 1000) + (Habitaciones * 5000) + Ruido_Aleatorio.

* 2) Visualiza: Haz un scatterplot de Metros vs Precio.

* 3) ML: Entrena un modelo que use Metros_Cuadrados para predecir el Precio.

* 4) Uso: Predice cu√°nto costar√≠a una casa de 120 m¬≤.

```python
#Generar el DataFrame de 100 casas
n=100
m2=np.random.randint(50,200, n)
habitaciones=np.random.randint(1,5,n)
ruido=np.random.randint(1,20000,n)
precio=(m2*1000)+(5000*habitaciones)+ruido

df_casas=pd.DataFrame({'Metros_cuadrados':m2,
                       'Habitaciones':habitaciones,
                       'Precio':precio})

df_casas.head()
```

```python
#Scatterplot metros vs precio
sns.scatterplot(df_casas, x='Metros_cuadrados', y='Precio')
```

```python
#Modelo que usa m2 para predecir precios
modelo=LinearRegression()
X=df_casas[['Metros_cuadrados']]
y=df_casas[['Precio']]
modelo.fit(X,y)
#Prediccion para 120 m2
m2_predecir=120
prediccion=modelo.predict([[m2_predecir]])

print(f'El precio para {m2_predecir} es ${prediccion[0][0]:.2f}')
```
