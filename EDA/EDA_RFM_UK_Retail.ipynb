```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset real de transacciones online (UCI Machine Learning Repository)
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv"

df_retail = pd.read_csv(url, encoding='ISO-8859-1')
df_retail['InvoiceDate'] = pd.to_datetime(df_retail['InvoiceDate'])
```

```python
df_sin_negativos=df_retail[(df_retail['Quantity']>0)&(df_retail['UnitPrice']>0)]
df_sin_negativos.describe()
```

```python
df_sin_negativos['TotalSum']=df_sin_negativos['Quantity']*df_sin_negativos['UnitPrice']
df_sin_negativos
```

```python
snapshot_date = df_sin_negativos['InvoiceDate'].max() + pd.Timedelta(days=1)
snapshot_date
```

FASE 2: Creando la Tabla de Clientes (RFM) 游논

Ahora mismo tienes filas de "productos". Necesitamos resumir esto a una fila por "Cliente" (CustomerID).

Fecha de referencia: Crea una variable snapshot_date que sea la fecha m치xima del dataset + 1 d칤a. (Para simular que hacemos el an치lisis "hoy").

* snapshot_date = df_retail['InvoiceDate'].max() + pd.Timedelta(days=1)

Agrupaci칩n Maestra: Agrupa por CustomerID y calcula:

* Recency: D칤as entre snapshot_date y la 칰ltima compra del cliente (max de la fecha).

* Frequency: Conteo (count) de sus facturas 칰nicas (InvoiceNo). (Nota: Para simplificar, puedes contar filas, o usar nunique si te animas).

* Monetary: Suma (sum) de TotalSum.

```python
rfm=df_sin_negativos.groupby('CustomerID').agg({
    'InvoiceDate':lambda x: (snapshot_date - x.max()).days, #쮿ace cu치nto compr칩?
    'InvoiceNo':'nunique', #쮺u치ntas veces vino?
    'TotalSum':'sum' #쮺u치nto gast칩?
})

rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)
```

```python
rfm.head()
```

FASE 3: El Ranking (Quantiles) 游끥 [Herramienta Nueva]

Queremos etiquetar a los clientes del 1 al 4 seg칰n qu칠 tan buenos son. Usaremos pd.qcut(columna, q=4, labels=[...]).

```python
# Creamos un ranking para desempatar a los que tienen la misma cantidad de compras.
# method='first' significa: si empatan, el que aparece primero en la lista gana.
rfm['Frequency_Rank'] = rfm['Frequency'].rank(method='first')


rfm['R_label']=pd.qcut(rfm['Recency'],q=4,labels=[4,3,2,1])
rfm['F_label']=pd.qcut(rfm['Frequency_Rank'],q=4,labels=[1,2,3,4], duplicates='drop')
rfm['M_label']=pd.qcut(rfm['Monetary'],q=4,labels=[1,2,3,4])
rfm.head()
```

Filtra y mu칠strame a los mejores clientes posibles (aquellos que tengan Score m치ximo, es decir, 4+4+4 = 12, o cercano).

```python
#Nos aseguramos que sean enteros
rfm['RFM_score']=rfm['R_label'].astype(int)+rfm['F_label'].astype(int)+rfm['M_label'].astype(int)
rfm.sort_values(by='RFM_score',ascending=False)
```
