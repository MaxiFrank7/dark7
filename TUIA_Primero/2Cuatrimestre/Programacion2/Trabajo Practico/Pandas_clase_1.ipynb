Comencemos leyendo el archivo con csv. Como podr치n recordar, extraer informaci칩n puede ser un poco tedioso ya que para poder obtener todos los valores en una columna, tenemos que recorrer todas las filas del archivo csv.

En este caso vamos a recolectar los precios y los precios clasificados por barrios

```python
import csv
lista_precios = []
precio_por_barrio = {}

with open('listings.csv') as File:
    next(File)
    reader = csv.reader(File, delimiter=',')
    for fila in reader:
        precio = float(fila[9].replace('$', '').replace(',',''))
        lista_precios.append(precio)
        #por barrio
        barrio = fila[5] # neighbourhood
        if barrio in precio_por_barrio:
            precio_por_barrio[barrio].append(precio)
        else:
            precio_por_barrio[barrio] = [precio]
```

```python
precio_por_barrio
```

```python
suma = 0
contador = 0
for precio in lista_precios:
  suma += precio
  contador += 1
suma / contador
```

```python
promedio_por_barrio = {}
for barrio, precio in precio_por_barrio.items():
  suma = 0
  for i in precio:
    suma += i
  promedio_por_barrio[barrio] = suma / len(precio)
```

```python
promedio_por_barrio
```

En el c칩digo de arriba generamos una lista con todos los precios que aparecen en el archivo `listing.csv` y un diccionario donde las claves son los barrios y los valores son todos los precios de ese barrio. A continuaci칩n hacemos una breve exploraci칩n de esos datos

# Tipos de datos
En la materia Programaci칩n 1 ya vimos que, en Python, los tipos b치sicos se dividen en:

* **N칰meros:**
  *   int (entero)
  *   float (de punto flotante)

En Python, los n칰meros enteros son exactos y de precisi칩n arbitraria, al  contrario de otros lenguajes como C o Java donde no se pueden representar n칰meros enteros demasiado grandes. En muchos lenguajes el tipo de dato int es fijo y hay limites a lo que puede almacenar. Si se sobrepasan se produce un desbordamiento u overflow. En Python no existen limites para el entero, si se necesita m치s memoria se reserva la misma seg칰n lo que se necesite (el l칤mite viene dado por la cantidad de memoria disponible).

Los tipos de datos de Pandas son una parte esencial para la manipulaci칩n y el an치lisis de datos en Python. Estos definen c칩mo se almacena y procesa la informaci칩n, **impactando tanto en la eficiencia como en la precisi칩n**. Este tema explorar치 los diversos tipos de datos en Pandas, incluyendo su importancia, relevancia para el an치lisis de datos y su relaci칩n con los tipos de datos de NumPy y de Python est치ndar.

En Pandas, los tipos de datos son una parte crucial de c칩mo se representa la informaci칩n y se clasifican en tres grupos principales: num칠ricos, categ칩ricos y de fecha y hora.

**Tipos num칠ricos**: incluyen enteros y n칰meros de punto flotante, que representan datos cuantitativos. Permiten realizar operaciones matem치ticas y son fundamentales para el an치lisis num칠rico.

**Tipos categ칩ricos**: representan datos cualitativos, como categor칤as o etiquetas. Al definir los datos como categ칩ricos, se mejora la eficiencia y se les da un significado m치s claro.

**Tipos de fecha y hora**: el manejo de fechas y horas puede ser complejo, pero Pandas ofrece tipos de datos espec칤ficos que facilitan esta tarea. Estos tipos permiten realizar ajustes de formato, de zona horaria y otras operaciones relacionadas con el tiempo.

Al trabajar con Pandas, es fundamental reconocer c칩mo estos tipos se alinean con los **tipos de datos nativos de Python y de NumPy**. Los tipos num칠ricos en pandas corresponden a los tipos `int` y `float` de Python, as칤 como a `int64` y `float64` de NumPy. Los tipos categ칩ricos pueden relacionarse con el tipo `str` o `list` de Python y con el tipo `object` de NumPy. Los tipos de fecha y hora tienen paralelos en el m칩dulo `datetime` de Python y en `datetime64` de NumPy.
Comprender las relaciones entre los tipos de datos en pandas, NumPy y los tipos nativos de Python permite transiciones m치s fluidas entre ellos.

#Salteando (skiping)
Puede ser que en algunas ocasiones, tal vez como la presentada anteriormente, querramos simplemente no trabajar con las filas que tengan datos de este tipo, o saltarlas por otras razones (no son relevantes para el an치lisis, o no son 칰tiles). Con esto podemos evitar errores, y adem치s al ser posible hacerlo cuando estamos leyendo el archivo puede resultar realmente 칰til cuando estamos trabajando con archivos grandes que puedan ralentizar significativamente el trabajo con ellos.

Entre los argumentos de `pd.read_csv` podemos encontrar `skiprows` donde se puede indicar un n칰mero de l칤neas a omitir, una lista con l칤neas a omitir, o un Callable (una funci칩n que indicar칤a dado alg칰n argumento qu칠 linea se omite). En caso de los Callables esta funci칩n se evaluar치 en los 칤ndices de las filas, retornando True si la fila debe omitirse y False en caso contrario. Un ejemplo de un argumento callable v치lido ser칤a `lambda x: x in [0, 2]`.

Otros argumentos relevantes son `skipfooter` de tipo int que nos permite eliminar un n칰mero espec칤fico de lineas desde el final del archivo (por defecto 0) y `header` que nos permite elegir que n칰mero de fila vamos a utilizar como nombres.

Otro argumento 칰til es `skipinitialspace` que nos permite para saltear espacios luego del delimitador (el separador de campos)

Probemos algunos de estos comandos:

```python
# Ejemplo  1: Saltar las filas 0 y 2 (usando un callable)
df_skip_rows = pd.read_csv("listings.csv", skiprows=lambda x: x in [0, 2])
print("DataFrame con filas 0 y 2 omitidas:")
print(df_skip_rows)

# Ejemplo 2: Saltar las 칰ltimas 5 filas
df_skip_footer = pd.read_csv("listings.csv", skipfooter=5, engine='python')
print("\nDataFrame con las 칰ltimas 5 filas omitidas:")
print(df_skip_footer)

# Ejemplo 3: Seleccionar una fila espec칤fica como header
df_header = pd.read_csv("listings.csv", header=1)  # Aqu칤 estamos usando la fila 1 como encabezado
print("\nDataFrame con la fila 1 como encabezado:")
print(df_header)

# Ejemplo 4: Saltar espacios en blanco despu칠s de los delimitadores
df_skip_space = pd.read_csv("listings.csv", skipinitialspace=True)
print("\nDataFrame ignorando espacios en blanco despu칠s del delimitador:")
print(df_skip_space)
```
