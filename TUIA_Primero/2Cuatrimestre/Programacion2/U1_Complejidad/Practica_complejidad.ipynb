### **Ejercicio 0**

```python
def condicional0(num: int) -> int:
    if num > 10:            #1 operacion O(1)                    
        print(f"{num} es mayor que 10") #1 operacion O(1) 
        aux = num % 2  #2 operaciones O(1) 1 asignar 1 modulo                   
    else:                   #1 operacion O(1)            
        print(f"{num} es menor que 10") #1 operacion O(1) 
        aux = num // 2 #2 operaciones O(1) 1 asignar 1 division
    return aux #1 operacion O(1) 
#rama True 1+1+2+1(return)=5. Rama false 1+1+2+1=5
#Es constante O(1)
```

### **Ejercicio 1**

```python
def condicional1(num: int) -> int:
    if num > 10:                               #1 operacion O(1)
        print(f"{num} es mayor que 10")        #1 operacion O(1)
        aux = num % 2                     #2 operaciones O(1) 
        if aux == 0:                           #1 operacion O(1)
            print(f"{num} es divisible por 2") #1 operacion O(1) --> 1+1+2+1+1+return(1)=7
        else:                                  
            print(f"{num} no es divisible por 2") #1 operacion O(1)  
            aux = num * num * num   #3 operaciones(2 multi 1 asign) O(1) --> 1+1+2+1+1+3+return(1)=10
    else:                                
        print(f"{num} es menor que 10")  #1 operacion O(1)
        aux = num // 2              #2 operaciones O(1)
        producto = num * 2          #2 operaciones O(1)
        if producto % 3 == 0:            #2 operaciones(modulo+comparacion) O(1)
            print("producto es divisible por 3") #1 operacion O(1) -->1+1+2+2+2+1+return=10
    return aux                           #1 operacion O(1)
#Es constante O(1)
```

### **Ejercicio 2**

```python
def ciclo0(n: int) -> None:
    lista = list(range(n))  #n asignaciones O(n)
    m = n // 2                    #2 operaciones O(1)
    for i in range(0, n):              #n asignaciones O(n)
        print("inicio ciclo j")        #1
        for j in range(m):             #n/2 asignaciones O(n/2) -> for anidado n2/2
            print("inicio ciclo k")    #1
            for k in range(0, 3):      #3 asignaciones O(1) -> for anidado 3n2/2
                print(lista[k])        #1
#Total= n+2+n*n/2*3= 3/2 n2+n+2 -> O(n2)
```

### **Ejercicio 8**

La siguiente función recibe dos listas ordenadas y devuelve una sola lista ordenada que contiene los mismos elementos que la concatenación de las dos anteriores, pero en orden.

```python
def mezcla_de_listas(L1: list[int], L2: list[int]) -> list[int]:
    # precond: L1 y L2 estan ordenadas
    i, j = 0, 0     #2
    L3 = []                      #1
    while i < len(L1) and j < len(L2):      #2 comparaciones que se hacen maximo l1+l2-1 veces : O(l1+l2)
        if L1[i] < L2[j]:                   #1
            L3.append(L1[i])                #1
            i += 1                          #2
        else:                               
            L3.append(L2[j])                #1
            j += 1                          #2
    if i == len(L1):                        #1
        for k in range(j, len(L2)):         #x distancia entre L2 y j: O(l2-j)
            L3.append(L2[k])                #1
    else:                                   
        for k in range(i, len(L1)):         #y distancia entre L1 e i: O(l1-i)
            L3.append(L1[k])                #1
    return L3                               #1

#Total true while=(L1+L2)*(2+1+1+2)=6(l1+l2) ->O(l1+l2)
#Total else while=(L1+L2)*(2+1+1+2)=6(l1+l2) ->O(l1+l2)
#Total if true=1+x*1= (L2-j) -> O(L2)
#Total else=1+y*1= (L1-i) ->O(L1)
#El mayor coste es O(l1+l2)
L1=[0,2,4,6,8,10]
L2=[1,3,5,7,9]
print(mezcla_de_listas(L1,L2))
```

### **Ejercicio 10**
La siguiente función determina si un número es primo:

```python
def primo(numero: int) -> bool:
    # Precond: numero es no negativo
    if numero <= 1:         #1
        return False        #1
    else:
        cont = 0            #1
        for i in range(2, numero + 1): #n-1 veces: O(n)
            if numero % i == 0: #2
                cont += 1       #2
        if cont == 1:           #1
            return True         #1
        else:
            return False        #1
#Total rama else: 1+1+(n-1)*(2+2)+2=4+4n-4=4n -> O(n)
```

### **Ejercicio 11**
La siguiente función calcula el factorial de forma iterativa:

```python
def factorial(numero: int):
    fact = 1        #1
    for i in range(1, numero + 1): #n
        fact *= 1   #2
    return fact     #1
#Total=1+n*2+1=2n+2 -> O(n)
```
