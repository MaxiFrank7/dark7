```python
class Point:
    """ representación de un punto en un plano cartesiano 2D """
    def __init__(self, x: float, y: float) -> None:
        self.x = x
        self.y = y
    def __str__(self) -> str:
        return '('+ str(self.x) + ', '+ str(self.y) + ')'
    def __eq__(self, other) -> bool:
        if not isinstance(other, Point):
            return NotImplemented
        return self.x == other.x and self.y == other.y
    def __add__(self, other: 'Point') -> 'Point':
        return Point(self.x + other.x, self.y + other.y)

class Rectangle:
    def __init__(self, width: float, height: float, corner: Point) -> None:
        self.width = width
        self.height = height
        self.corner = corner
    def __str__(self)-> str:
        return f"base: {str(self.width)}, altura: {str(self.height)} en esquina {str(self.corner)}"
    def __eq__(self, other:'Rectangle') -> bool:
        if not isinstance(other, Rectangle):
            return NotImplemented
        return self.height==other.height and self.width==other.width and self.corner.__eq__(other.corner) #como esta definido el eq en point se podria usar self.corner==other.corner

def mover_rectangulo_modificadora(rectangulo:Rectangle, dx:float, dy:float )-> None:
    rectangulo.corner.x+=dx
    rectangulo.corner.y+=dy
def mover_rectangulo_pura(rectangulo:Rectangle, dx:float, dy:float)-> Rectangle:
    import copy
    r2=copy.deepcopy(rectangulo)
    r2.corner.x+=dx
    r2.corner.y+=dy
    return r2

p1=Point(3,4)
print(p1.__str__())
r1=Rectangle(10,20,p1)
print(r1.__str__())
mover_rectangulo_modificadora(r1,5,5)
print(r1.__str__())
r2=mover_rectangulo_pura(r1,10,10)
print(r2)
```

### **Ejercicio 2**

Defina, en la clase Point, un método distancia() que nos de la distancia euclídea entre dos puntos.

Nota: Recordar que la distancia euclídea entre dos puntos A = (x1, y1) y B = (x2, y2) se calcula como: d = sqrt((x2−x1)2 + (y2−y1)2) 

```python
class Point:
    def __init__(self, x:float, y:float)->None:
        self.x=x
        self.y=y
    def distancia(self, other: 'Point') -> float:
        d=((self.x-other.x)**2+(self.y-other.y)**2)**0.5
        return d
p1=Point(1,2)
p2=Point(2,3)
print(f'La distancia entre p1 y p2 es {p1.distancia(p2)}')
        
```

### **Ejercicio 4**

Defina una clase Robot que simule los movimientos de un robot y calcule la posición en la que se encuentra cada momento. El robot se moverá por un tablero infinito de coordenadas X e Y, podrá realizar los siguientes movimientos:
* Avanzar hacia adelante (W).
* Retroceder (S).
* Avanzar hacia la izquierda (A) o hacia la derecha (D).

El robot tendrá un método llamado mueve() que recibirá la orden como parámetro y otro método, posicion_actual(), que indicará su posición en las coordenadas X e Y. Al crear el robot, este se inicializará a las coordenadas (0, 0).


```python
class Robot:
    def __init__(self, x:int=0, y:int=0) -> None:
        self.x=x
        self.y=y
    def mueve(self, orden:str) -> None:
        orden=orden.lower()
        if orden=='w':
            self.y+=1
        elif orden == 's':
            self.y-=1
        elif orden == 'a':
            self.x-=1
        elif orden == 'd':
            self.x+=1
        else:
            print(f'Orden invalida, por favor ingrese W/A/S/D: ')
    def posicion_actual(self)->str:
        return f'El robot esta en la coordenada({self.x},{self.y})'
    
mi_robot = Robot()
orden = input("Introduce la orden (W/A/S/D o fin para finalizar): ")
while orden != 'fin':
    mi_robot.mueve(orden)
    print(mi_robot.posicion_actual())
    orden = input("Introduce la orden: ")
        
```

* Agregar un método como_volver que indique la secuencia de movimientos que deberíamos realizar
para regresar a la coordenada (0, 0)

```python
class Robot:
    def __init__(self, x:int=0, y:int=0) -> None:
        self.x=x
        self.y=y
        self.historial: list[str] = []
    
    def mueve(self, orden:str)-> None:
        orden=orden.upper()
        lista=list(orden)
        for letra in lista:
            if letra not in 'WASD':
                print(f'Secuencia de movimientos no valido! Ingrese solamente W/A/S/D')
                return
        for letra in lista:
            if letra=='W':
                self.y+=1
            elif letra == 'S':
                self.y-=1
            elif letra == 'A':
                self.x-=1
            elif letra == 'D':
                self.x+=1
            self.historial.append(letra)
    
    def posicion_actual(self)->str:
        return f'El robot esta en la coordenada({self.x},{self.y})'
    
    def obtener_historico_de_movimientos(self)->str:
        return f'Historial de movimientos: {"".join(self.historial)}'
    
    def como_volver(self)->str:
        vuelta:list[str]=[]
        if self.x>0:
            vuelta.append('A'*self.x)
        elif self.x<0:
            vuelta.append('D'*(-self.x))
        
        if self.y>0:
            vuelta.append('S'*self.y)
        elif self.y<0:
            vuelta.append('W'*(-self.y))
        return f'Para volver en la menor cantidad de moviemientos: {"".join(vuelta)}'

mi_robot = Robot()
orden = "WSADADDSASD"
mi_robot.mueve(orden)
print(mi_robot.posicion_actual())
print(mi_robot.obtener_historico_de_movimientos())
print(mi_robot.como_volver())
```

### **Ejercicio 6**

Crear las clases Materia y Carrera.
Cada materia tiene:
* Un código de materia.
* Un nombre de la materia.
* Una cantidad de créditos que aporta.
* Una carrera puede pensarse como una lista de materias.

La clase debería comportarse de forma que el siguiente ejemplo tenga sentido:
* analisis2 = Materia("61.03", "Análisis 2", 8)
* fisica2 = Materia("62.01", "Física 2", 8)
* algo1 = Materia("75.40", "Algoritmos 1", 6)
* c = Carrera([analisis2, fisica2, algo1])
* print(c)
    * #3Créditos: 0 -- Promedio: N/A -- Materias aprobadas:
* c.aprobar("95.14", 7)
    * #Error: La materia 75.14 no es parte del plan de estudios
* c.aprobar("75.40", 10)
* c.aprobar("62.01", 7)
* print(c)
    * #Créditos: 14 -- Promedio: 8.5 -- Materias aprobadas: 75.40 Algoritmos 1 (10) 62.01 Física 2 (7)

```python
class Materia:
    """Representa una materia del plan de estudios."""
    def __init__(self, codigo:str, nombre:str, creditos:int) -> None:
        self.codigo=codigo
        self.nombre=nombre
        self.creditos=creditos
    def __str__(self) -> str:
        return f'Nombre: {self.nombre}, código: {self.codigo}, créditos: {self.creditos}'

class Carrera:
    """Administra el plan de materias y el progreso del estudiante."""
    def __init__(self, materias:list[Materia]) -> None:
        self.plan:dict[str, Materia] = {} #Diccionario índice por código para buscar materias rápido
        for m in materias:
            self.plan[m.codigo] = m
        self.aprobadas:dict[str,int]={} #Diccionario de materias aprobadas, código -> nota
    
    def aprobar(self, codigo:str, nota:int) -> None:
        """Registra que una materia del plan fue aprobada con una nota."""
        if codigo in self.plan:
            self.aprobadas[codigo]=nota
        else:
            print(f'Error: La materia {codigo} no es parte del plan de estudios')
            return
    def creditos(self)->int:
        """Suma los créditos de todas las materias aprobadas."""
        suma=0
        for codigo in self.aprobadas:
            suma+=self.plan[codigo].creditos  #Tomamos los créditos desde el plan usando el código
        return suma
    def promedio(self)->float|None:
        """Calcula el promedio de notas, si no hay aprobadas devuelve None."""
        suma=0
        cont=0
        for codigo, nota in self.aprobadas.items():
            suma+=nota
            cont+=1
        if cont==0:
            return None
        else:
            return round(suma/cont,2)
        
    def __str__(self)->str:
        """Devuelve un resumen del estado de la carrera."""
        if not self.aprobadas: #Caso sin materias aprobadas
            return f'Créditos: 0 -- Promedio: N/A -- Materias aprobadas:' 
        lista_aprobadas=[]
        for codigo in self.aprobadas:
            nombre=self.plan[codigo].nombre
            nota=self.aprobadas[codigo]
            lista_aprobadas.append(f'{codigo} {nombre} ({nota}) ')
        texto_aprobadas="".join(lista_aprobadas) #Unimos todos los textos en una sola cadena
        return f'Créditos: {self.creditos()} -- Promedio: {self.promedio()} -- Materias aprobadas: {texto_aprobadas}'

#TESTEO
analisis2 = Materia("61.03", "Análisis 2", 8)
fisica2 = Materia("62.01", "Física 2", 8)
algo1 = Materia("75.40", "Algoritmos 1", 6)

print(analisis2)
print(fisica2)
print(algo1)
c=Carrera([analisis2, fisica2, algo1])
print(c)
c.aprobar("95.14", 7)
c.aprobar("75.40", 10)
c.aprobar("62.01", 7)
print(c)
        
        
```

Complete la funcionalidad de la clase Jugador, implementando los siguientes métodos:

    * golpeado: quita vida al jugador.
    * golpear: quita vida al enemigo y lo agrega a la lista de enemigos golpeados.


```python
class Entidad:
    def __init__(self, vida_inicial: int):
        self.vida = vida_inicial
class Enemigo(Entidad):
    pass
class Jugador(Entidad):
    def __init__(self, vida_inicial: int):
        super().__init__(vida_inicial)
        self.enemigos_golpeados = []
    def golpeado(self, cuanto: int):
        """Cuanto de vida perdio el jugador."""
        self.vida-=cuanto
    def golpear(self, enemigo: Enemigo, cuanto: int):
        """quita vida al enemigo y lo agrega a la lista de golpeados."""
        enemigo.vida-=cuanto
        if enemigo not in self.enemigos_golpeados:
            self.enemigos_golpeados.append(enemigo)
```
