### **Ejercicio 0**

Escriba una función recursiva factorial que tome un numero natural n y calcule su factorial n!

```python
def factorial_recursiva(n: int) -> int:
    if n == 0:
        return 1
    return n * factorial_recursiva(n - 1)

print(factorial_recursiva(4))
```

### **Ejercicio 2**
Escriba una función recursiva fibonacci que calcule el n-ésimo número de Fibonacci

```python
def fibo(n:int)->int:
    if n==0:
        return 0
    if n==1:
        return 1
    return fibo(n-1)+fibo(n-2)
print(fibo(8))
```

### **Ejercicio 3**
1. Escriba una función recursiva print_string que reciba como parámetro un número natural n y
una string s y escriba por pantalla n veces el valor de la string s.

```python
def print_string(n:int, saludo:str) -> None:
    if n<=0:
        return
    print(saludo)
    print_string(n-1,saludo)

cadena = "Test"
print_string(3, cadena)
```

2. Escriba una función recursiva repeat_string que reciba como parámetro un número natural n y
una string s y devuelva el valor de n concatenaciones de la string s.

```python
def repeat_string(n: int, saludo: str) -> str:
    if n <= 0:
        return ""
    return saludo + repeat_string(n - 1, saludo)


cadena = "Test"
print(repeat_string(3, cadena))
```

### **Ejercicio 4**
Convierta la siguiente función iterativa a una definición recursiva.

```python
def iterativa(l: list[int]) -> int:
    c = 1
    for i in l:
        c = c * i
    return c
print(iterativa([1864654,-1]))
```

```python
def recursiva(l: list[int]) -> int:
    if len(l)==0:
        return 1
    return l[0]*recursiva(l[1:])

print(recursiva([1864654,-1]))
```

### **Ejercicio 5**
Escriba una función recursiva find_max que encuentre el mayor elemento de una lista de números
naturales.

```python
def find_max(numeros:list[int])->int:
    if len(numeros)==0:
        return 0
    if len(numeros)==1:
        return numeros[0]
    maximo= find_max(numeros[1:])
    if numeros[0] > maximo:
        return numeros[0]
    else:
        return maximo

print(find_max([1,3,2]))
```

```python
def power(a:int, b:int) -> int:
    if b==0:
        return 1
    return a*power(a,b-1)

print(power(2,3))
```

### **Ejercicio 9**
Escriba una función recursiva replicate(l, n) que replique n veces cada elemento en la lista l.

```python
def replicate(l:list[int],n:int)->list[int]:
    if len(l)==0:
        return []
    return [l[0]]*n+replicate(l[1:],n)

print(replicate([1,3,3,7],2))
```

### **Ejercicio 10**
Escriba una función recursiva is_palindrome que compute si una string es un palíndromo.
* Nota: Un palíndromo es una secuencia de caracteres que se lee igual hacia adelante que hacia atrás.
* > is_palindrome("") -> True
* > is_palindrome("a") -> True
* > is_palindrome("ejemplo") -> False
* > is_palindrome("sometemos") -> True


```python
def is_palindrome(cadena:str)->bool:
    if len(cadena)==0:
        return True
    if len(cadena)<=1:
        return True
    if cadena[-1]!=cadena[0]:
        return False
    return is_palindrome(cadena[1:-1]) #verifica cadena interior por ej en "sometemos" verifica "ometemo" y asi

print(is_palindrome("a"))
print(is_palindrome("ejemplo"))
print(is_palindrome("sometemos"))
```

### **Ejercicio 12**
Escriba una función recursiva potencia(a: int, b: int) que reciba dos números enteros (incluyendo negativos) a y b y calcule la potencia a^b.

Para esto, reutilice la función power definida en el Ejercicio 6.

```python
def power(a:int, b:int) -> int:
    if b==0:
        return 1
    return a*power(a,b-1)

def potencia(a:int, b:int)->int|float:
    if b>=0:
        return power(a,b)
    if b<0:
        return 1/float(power(a,-b))

print(potencia(2,-4))
```

### **Ejercicio 13**
Escriba una función average que calcule el promedio de una lista de números. Para esto, defina una función recursiva auxiliar aux que calcule tanto la cantidad y la suma de los elementos de la lista.

```python
def aux(lista: list[float]) -> tuple[int, float]:
    if len(lista)==0:
        return (0,0.0)
    if len(lista)==1:
        return (1,lista[0])
    cont, suma=aux(lista[1:])
    return (1+cont, lista[0]+suma)

print(aux([10.5,100.0,150.5]))

def average(lista: list[float]) -> float|None:
    if len(lista)==0:
        return 0.0
    cant,suma=aux(lista)
    return suma/cant

print(average([10.5,100.0,150.5]))
```

### **Ejercicio 14**
Escribir una función recursiva que reciba como parámetros dos cadenas a y b, y encuentre la posición de la primer ocurrencia de b como subcadena de a.
Ejemplo:
* > posicion("Wubba Lubba Dub Dub!", "ubba") -> 1
* > posicion("Ñam fri fruli fali fru", "fru") -> 8

Ayuda: Puede ser de utilidad el método de str.startswith que nos dice si una cadena empieza con una subcadena dada. Puede leer más sobre este método en la documentación oficial.

```python
def posicion(a: str, b: str) -> (int|None):
    if len(a) == 0:
        return None
    if a.startswith(b):
        return 0
    if posicion(a[1:], b) is None:
        return None
    return 1 + posicion(a[1:], b)

print(posicion("Wubba Lubba Dub Dub!", "ubba"))

print(posicion("Usted es el Jefe de los Minisuper?", "Si"))
```
