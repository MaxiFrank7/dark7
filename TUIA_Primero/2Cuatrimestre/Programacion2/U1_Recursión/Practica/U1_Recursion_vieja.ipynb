### **Ejercicio 0**

Escriba una función recursiva factorial que tome un numero natural n y calcule su factorial n!

```python
def factorial_recursiva(n: int) -> int:
    if n == 0:
        return 1
    return n * factorial_recursiva(n - 1)

print(factorial_recursiva(4))
```

### **Ejercicio 1**
1. Escriba una función recursiva repite_hola que reciba como parámetro un número entero n y
escriba por pantalla n veces el mensaje "Hola". Invóquela con distintos valores de n.

```python
def repite_hola(n: int) -> None:
    if n <= 0:
        print("")
    for _ in range(n):
        print("Hola")


repite_hola(7)
```

```python
def repite_hola2(n: int) -> str:
    if n <= 0:
        return ""
    return "Hola" + repite_hola2(n - 1)


print(repite_hola2(3))
```

### **Ejercicio 3**
1. Piense cuál sería el resultado de la expresión misterio(5) y luego compruebe su hipótesis
ejecutándola.
2. Explique con sus palabras qué hace misterio(a) para cualquier a, sea positivo, negativo o 0.

```python
def misterio(a: int) -> int:
    if a == 0:
        return a
    return 1 + misterio(a - 1)


print(misterio(5))
```

* Si a=0, devuelve 0. 
* Si a es positivo la función hace una recursión decreciendo de uno en uno hasta llegar a 0, sumando un 1 en cada paso.
* Si a es negativo, nunca llega al caso base porque se vuelve cada vez mas negativo, es decir se aleja del caso base. **Recursion infinita**

***Resolviendo misterio(5)***


* misterio(5) → como a != 0, devuelve 1 + misterio(4)
* misterio(4) → devuelve 1 + misterio(3)
* misterio(3) → devuelve 1 + misterio(2)
* misterio(2) → devuelve 1 + misterio(1)
* misterio(1) → devuelve 1 + misterio(0)
* misterio(0) → ahora sí, como a == 0, devuelve 0

***Reemplazando hacia arriba:***

* misterio(0) = 0
* misterio(1) = 1 + 0 = 1
* misterio(2) = 1 + 1 = 2
* misterio(3) = 1 + 2 = 3
* misterio(4) = 1 + 3 = 4
* misterio(5) = 1 + 4 = 5


```python
def misterio(a: int) -> int:
    if a == 0:
        return a
    elif a > 0:
        return 1 + misterio(a - 1)
    else:  # caso negativo
        return -1 + misterio(a + 1)


print(misterio(5))
```

* misterio(-3) = -1 + misterio(-2)
* = -1 + (-1 + misterio(-1))
* = -1 + (-1 + (-1 + misterio(0)))
* = -1 + (-1 + (-1 + 0))
* = -3

### **Ejercicio 4**
Considere la siguiente función recursiva:

```python
def f(n: int, d: int) -> None:
    if n > 1:
        if n % d == 0:
            print(d)
            f(n // d, d)
        else:
            f(n, d + 1)
f(30,2)
```

1. ¿Qué muestra por pantalla la llamada f(30, 2)? Intente deducirlo sin ejecutar el código.
2. A nivel general, ¿qué muestra por pantalla la llamada f(x, 2) para un x cualquiera?
3. Implemente una función iterativa equivalente.

1. * En f(30,2) primero printea 2 
   * y luego llama recursivamente a f(15,2) como 15%2 != 0 va al else 
   * y llama recursivamente f(15,3) y va a printear 3 
   * y llama recursivamente f(5,3) como 5%3 != 0 va al else
   * y llama recursivamente f(5,4) como 5%4 != 0 va al else
   * y llama recursivamente f(5,5) printea 5
2. Va a mostrar en pantalla todos los primos desde X hasta el 2

```python
def ff(n: int, d: int) -> None:
    while n>=d:
        if n%d==0:
            print(d)
            n=n//d
        else:
            d+=1
ff(30,2)
```

### **Ejercicio 7**
Convierta la siguiente función recursiva a una iterativa.

```python
def recursiva(t: int, k: int) -> int:
    if t >= 100:
        return k
    else:
        return recursiva(t + k, k + 1)

print(recursiva(10,2))
```

```python
def iterativa(t: int, k: int) -> int:
    while t<100:
        t=t+k
        k+=1
    return k
print(iterativa(10,2))
```

```python
def es_potencia(n:int, b:int)->bool:
    if n==1:
        return True
    elif n%b!=0 or n<1 or b<2:
        return False
    return es_potencia(n//b,b)
print(es_potencia(81,1)) 
```

## **Recursión con listas**
### **Ejercicio 11**
Escriba una función recursiva que encuentre el mayor elemento de una lista.

```python
def maximo_lista(numeros:list)->int:
    if len(numeros)==0:
        return 0
    if len(numeros)==1:
        return numeros[0]
    maximo= maximo_lista(numeros[1:])
    if numeros[0] > maximo:
        return numeros[0]
    else:
        return maximo

print(maximo_lista([]))
```

### **Ejercicio 12**
Convierta la siguiente función iterativa a una recursiva.

```python
def iterativa(l: list[int]) -> int:
    c = 1
    for i in l:
        c = c * i
    return c
print(iterativa([]))
```

```python
def recursiva(l: list[int]) -> int:
    if len(l)==0:
        return 1
    return l[0]*recursiva(l[1:])

print(recursiva([1864654,-1]))
```

### **Ejercicio 13**
Escriba una función recursiva para replicar los elementos de una lista una cantidad n de veces. Por ejemplo:

• replicar([1, 3, 3, 7], 2) -> ([1, 1, 3, 3, 3, 3, 7, 7])

```python
def replicar(l:list,n:int)->list:
    if len(l)==0:
        return []
    return [l[0]]*n+replicar(l[1:],n)
print(replicar([1,3,3,7],2))
```

### **Ejercicio 14**
Escriba una función que tome una lista y devuelva esa misma lista en orden inverso. Realice dos versiones:
* reversaR que resuelva utilizando recursión.
* reversaI que resuelva utilizando iteración.

Nota: No utilice la función built-in reversed en su solución, ni el método reversed.

```python
def reversaR(l:list)->list:
    if len(l)==0:
        return []
    return reversaR(l[1:])+[l[0]]
print(reversaR([1,2,3]))
```

```python
def reversaI(l:list)->list:
    salida=[]
    for i in range(len(l)-1,-1,-1):
        salida.append(l[i])
    return salida
print(reversaI([1,2,3]))
```

## **Recursión con strings**
### **Ejercicio 15**
Escribir una función recursiva que reciba como parámetros dos cadenas a y b, y encuentre la posición de la primer ocurrencia de b como subcadena de a.
Ejemplo:
* posicion_de("Un tete a tete con Tete", "te") -> 3
* posicion_de("Un tete a tete con Tete", "Te") -> 19

Ayuda: Puede ser de utilidad el método de str.startswith que nos dice si una cadena empieza con una subcadena dada. Puede leer más sobre este método en la documentación oficial.

```python
def posicion_de(a: str, b: str) -> int:
    if len(a) == 0:
        return -1
    if a.startswith(b):
        return 0
    if posicion_de(a[1:], b) == -1:
        return -1
    return 1 + posicion_de(a[1:], b)

print(posicion_de("Un tete a tete con Tete", "Te"))
```
